# Structured Output

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str

agent = create_agent(
    model='gpt-5-nano',
    system_prompt="You are a science fiction writer, create a space capital city at the users request",
    response_format = CapitalInfo
)

question = HumanMessage(content="What is Ares, Mars number one rival?")

response = agent.invoke(
    {"messages": [question]}
)

response["structured_response"]
```

```python
response["structured_response"].name
```

```python
capital_info = response["structured_response"]

capital_name = capital_info.name
capital_location = capital_info.location

print(f"{capital_name} is a city located at {capital_location}")
```

# Tool definition

```python
from langchain.tools import tool

@tool
def square_root(x: float) -> float:
    """Calculate the square root of a number."""
    return x ** 0.5 
```

```python
@tool('square_root') #overwrite tool name in decorator
def tool1(x: float) -> float:
    """Calculate the square root of a number."""
    return x**0.5
```

```python
@tool('square_root', description='Square root of a number') #overwrite tool description
def tool1(x: float) -> float:
    return x ** 0.5
```

```python
tool1.invoke({'x': 467})
```

# Adding to agents

```python
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5-nano",
    tools=[tool1],
    system_prompt="You are an arithmetic wizard. Use your tools to calculate the square root and square of any number."
)
```

```python
from langchain.messages import HumanMessage
question = HumanMessage(content="What is the square root of 467?")

response = agent.invoke({"messages": [question]})

print(response['messages'][-1].content)
```

# Wedding Planner Agent

### Required:travel agent (flights to and from location),venue agent(find venue for the wedding), dj agent(playlist matching vibe), and coordinator agent(da boss)

```python
import langchain
import base64
import dotenv
from langchain.agents import create_agent
from langchain.chat_models import init_chat_model
from langchain.tools import tool
from langchain.messages import HumanMessage
from langchain.messages import AIMessage
from pydantic import BaseModel
from pprint import pprint
from langchain.agents import AgentState
from langgraph.checkpoint.memory import InMemorySaver
from typing import List, Dict, Any
from langgraph.types import Command
from langchain.messages import ToolMessage
from langchain.tools import ToolRuntime
from tavily import TavilyClient
# Initialize Clients
dotenv.load_dotenv()
tavily_client = TavilyClient()

#Agent Environment
system_prompt_dj="You are a world class DJ hired to find the perfect playlist for a avant garde wedding. Use the Web to find the perfect playlist for the wedding. Then update the state with the information."
system_prompt_travel="You are a world class travel agent hired to find the perfect destination for a avant garde wedding. Read the current state to find the venue, then use the web to find the perfect travel route for the wedding. Then update the state with the information."
system_prompt_venue="You are a world class venue agent hired to find the perfect venue for a avant garde wedding. Use the web to find the perfect venue for the wedding. Then update the state with the information. Then call the travel agent to run his responsibilities."
system_prompt="You are the coordinator for three other agents, a DJ, a travel agent, and a venue agent. Make sure they all work together to create a perfect avant garde wedding."
class WeddingState(AgentState):
    playlist: List[str]
    venue: str
    travel: str
@tool
def GetTableNames() -> Dict[str, Any]:
    """Get the table names."""
    return Command(read={"messages": [ToolMessage("Successfully read table names", tool_call_id=runtime.tool_call_id)]}
    )
@tool
def UpdateTravel(travel: str) -> Dict[str, Any]:
    """Update the travel."""
    return Command(update={"travel": travel,
    "messages": [ToolMessage("Successfully updated travel", tool_call_id=runtime.tool_call_id)]}
    )
@tool
def UpdatePlaylist(playlist: List[str]) -> Dict[str, Any]:
    """Update the playlist."""
    return Command(update={"playlist": playlist,
    "messages": [ToolMessage("Successfully updated playlist", tool_call_id=runtime.tool_call_id)]}
    )
@tool
def UpdateVenue(venue: str) -> Dict[str, Any]:
    """Update the venue."""
    return Command(update={"venue": venue,
    "messages": [ToolMessage("Successfully updated venue", tool_call_id=runtime.tool_call_id)]}
    )
@tool
def web_search(query: str) -> Dict[str, Any]:
    """Search the web for information."""
    return tavily_client.search(query)
@tool
def call_djagent(query: str) -> Dict[str, Any]:
    """Call the DJ agent."""
    return djagent.run(query)
@tool
def call_travelagent(query: str) -> Dict[str, Any]:
    """Call the travel agent."""
    return travelagent.run(query)
@tool
def call_venueagent(query: str) -> Dict[str, Any]:
    """Call the venue agent."""
    return venueagent.run(query)
@tool
def read_venue() -> Dict[str, Any]:
    """Read the venue."""
    try:
        return runtime.state["venue"]
    except KeyError:
        return "No venue found in state"
@tool
def read_playlist() -> Dict[str, Any]:
    """Read the playlist."""
    return Command(read={"messages": [ToolMessage("Successfully read playlist", tool_call_id=runtime.tool_call_id)]}
    )
@tool
def read_travel() -> Dict[str, Any]:
    """Read the travel."""
    return Command(read={"messages": [ToolMessage("Successfully read travel", tool_call_id=runtime.tool_call_id)]}
    )
#===========CREATE AGENT========================
model = init_chat_model(model="gpt-5-nano", temperature=1, timeout=60, max_retries=2)
agent = create_agent(
    model,
    system_prompt=system_prompt,
    tools=[call_djagent,call_venueagent,read_venue,read_playlist,read_travel],
    state_schema=WeddingState(),
    checkpointer=InMemorySaver(),
)
#===========CREATE SUB AGENTS========================
djagent = create_agent(
    model='gpt-5-nano',
    tools=[web_search,UpdatePlaylist,read_playlist,GetTableNames],
    system_prompt=system_prompt_dj,
    checkpointer=InMemorySaver(),
)
travelagent = create_agent(
    model='gpt-5-nano',
    tools=[web_search,UpdateTravel,read_travel,read_venue],
    system_prompt=system_prompt_travel,
    checkpointer=InMemorySaver(),
)
venueagent = create_agent(
    model='gpt-5-nano',
    tools=[web_search,UpdateVenue,read_venue,call_travelagent],
    system_prompt=system_prompt_venue,
    checkpointer=InMemorySaver(),
)
#======CREATE QUESTIONS=============
question="Please plan a perfect weddings for the particiapants take if they live in Cincinnati, OH."

#===============RESPONSES===============================
response = agent.invoke({"messages": [HumanMessage(content=question)]})
pprint(response)



#===========STREAM RESPONSE=====================
# print("Invoking agent...")

# # We pass the question directly in the messages list
# for token, metadata in agent.stream(
#     {"messages": [HumanMessage(content=initial_message), question]},
#     config={"configurable": {"thread_id": "wedding1"}}, # Added required checkpointer config
#     stream_mode="messages"
# ):
#     if token.content: #check actual content
#         print(token.content, end="", flush=True) #print token
```
