# Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me about Ares, the capital of Mars")]},
    stream_mode="messages"
):
    #metadata contains which node produced the token

    if token.content: #check actual content
        print(token.content, end="", flush=True) #print token
```

```python
system_prompt="You are a science fiction writer, create a capital city at the users request."
question = "Tell me about Ares, the capital of Mars"
scifi_agent = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt
)

response = scifi_agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

```python
# system prompts ARE finetuning strategies for LLMs
# you can build them to look like structure you want from a response: example
system_prompt = """
You are a science fiction writer, create a space capital city at the users request.

Please keep to the below structure.

Name: Name of the capital city

Location: Where it is based

Vibe: 2-3 words to describe the vibe of the city

Economy: Main industries

"""

scifi_agent = create_agent(
    model='gpt-5-nano',
    system_prompt=system_prompt
    )

response =scifi_agent.invoke(
    {"messages": [question]
})
print(response['messages'][1].content)
```

# Structured Output

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str

agent = create_agent(
    model='gpt-5-nano',
    system_prompt="You are a science fiction writer, create a space capital city at the users request",
    response_format = CapitalInfo
)

question = HumanMessage(content="What is Ares, Mars number one rival?")

response = agent.invoke(
    {"messages": [question]}
)

response["structured_response"]
```

```python
response["structured_response"].name
```

```python
capital_info = response["structured_response"]

capital_name = capital_info.name
capital_location = capital_info.location

print(f"{capital_name} is a city located at {capital_location}")
```
